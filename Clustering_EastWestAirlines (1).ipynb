```python
#heirarchical clustering
#import libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
#import dataset
df=pd.read_excel('D:\Assignment\Clustering\EastWestAirlines.xlsx',sheet_name = "data") 
```

```python
df.info()
```

```python
df2=df.drop(['ID#'],axis=1)
df2
```

```python
def norm_func(i):      #normalization
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df2_norm = norm_func(df2)
df2_norm
```

```python
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(df2_norm,'complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'ward')
```

```python
y=pd.DataFrame(hc.fit_predict(df2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
#K-means clustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
data = []
for i in range (1,11):
    kmeans= KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df2_norm)
    data.append(kmeans.inertia_)
plt.plot(range(1, 11), data)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Data')
plt.show()
```
